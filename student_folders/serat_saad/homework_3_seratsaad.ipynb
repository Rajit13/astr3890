The data in hw3_populations.txt (source: scipy-lectures.org) describes the populations of hares and lynxes (and carrots) in northern Canada during 20 years:


create a Python script to

1. plot hw3_populations.txt (using matlab)


2. calculate and output:

    a) The mean, median and std of the populations of each species for the years in the period.
    
    b) The year each species had the largest population.
    
    c) The species that has the largest population for each year. (Hint: argsort & fancy indexing of np.array(['H', 'L', 'C']))
    
    d) The years were any of the populations is above 50000. (Hint: comparisons and np.any)
    
    e) The top 2 years for each species when they had the lowest populations. (Hint: argsort, fancy indexing)
    
    f) Compare (plot) the change in hare population (see help(np.gradient)) and the number of lynxes. Check correlation (see help(np.corrcoef)).



### Solution



```python
#importing necessary tools
import numpy as np
from matplotlib import pyplot as plt

#importing data from the file
data = np.loadtxt('hw3_populations.txt')

#assigning variables
#the variables are years, hare, lynx, and carrot
years = data[:,0]
hares = data[:,1]
lynxes = data[:,2]
carrots = data[:,3]

#plotting the populations
plt.plot(years, hares, color = 'red',)
plt.plot(years, lynxes, color = 'green')
plt.plot(years, carrots, color = 'blue')
plt.ylabel("Number")
plt.xlabel("Year")
plt.title("Populations Plotting")
plt.show()


```

In the above graph,

Red color shows the population of Hares

Green color shows the population of Lynxes

Blue color shows the population of Carrots

```python
#defining common function for printing data(with one variable)
def print_data(x):
    print("Mean: " + str(np.average(x)))
    print("Median: " + str(np.median(x)))
    print("STD: " + str(np.std(x)))
    print("Highest Population Year: " + str(years[np.argmax(x)]))


#printing data for hares
print("Hares")
print_data(hares)
print()

#printing data for lynxes
print("Lynxes")
print_data(hares)
print()

#printing data for carrots
print("Carrots")
print_data(hares)
print()
```

```python
#printing the highest population in each of the years
pop_type = ['Hares', 'Lynxes', 'Carrots']
max_per_year = (np.argsort(data[:,1:4]))[:, -1]

for i in range(0, 21):
    print(int(years[i]),':', pop_type[max_per_year[i]])
```

```python
#printing the year when value of any if the population is greater than 50000
print("When the population of any of them is above 50000:")
print()
print(years[np.any(data[:,1:]>50000, axis=1)])
```

```python
#printing the two years when the population was highest for each of the population 
print("Highest 2 years for different populations:")
print("Hares: " + str(int(years[np.argsort(hares)[-2]])) + " & " + str(int(years[np.argsort(hares)[-1]])))
print("Lynxes: " + str(int(years[np.argsort(lynxes)[-2]])) + " & " + str(int(years[np.argsort(lynxes)[-1]])))
print("Carrots: " + str(int(years[np.argsort(carrots)[-2]])) + " & " + str(int(years[np.argsort(carrots)[-1]])))
```

```python
#plotting the change in hare populatiojnand lynx population
plt.scatter(lynxes, np.gradient(hares), color = 'red')
plt.xlabel('Change in Hare Population')
plt.ylabel('Lynx Population')
plt.title("Comparison")
plt.show()

np.corrcoef(np.gradient(hares), lynxes)
```

Here through the graph it is true and visible that at the time when the change of hare population increase, the Lynx population decrease, similarly when the change in Hare Population increase, the Lynx population decreases. Also, the correlation is saying that the two values(change in hare populaiton and lynx polpulation) are negatively correlated to each other.
