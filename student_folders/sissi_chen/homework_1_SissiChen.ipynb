```python
%matplotlib inline
%config InlineBackend.figure_format='retina'

import matplotlib.pyplot as plt
import numpy as np

x = np.random.uniform(low=0.1, high=10, size=1000)
plt.hist(x, bins = 100)
plt.show()

y = np.log10(x)
plt.hist(y, bins = 100)
plt.show()

print('mean of x:', x.mean())
print('mean of y:', y.mean())

# Find median
x1 = x[49]
x2 = x[50]
median_x = (x1+x2)/2
print('median of x:', median_x)

y1 = y[49]
y2 = y[50]
median_y = (y1+y2)/2
print('median of y:', median_y)
```
